### Abstract

The cybersecurity attack dataset consists of network traffic logs and security warnings recorded across many network segments. It contains 40,000 records and 25 attributes that describe significant aspects of cyberattacks, such as the type of attack, source and destination IP addresses, network protocols utilized, anomaly scores, and severity levels.

This dataset is an invaluable resource for performing exploratory data analysis (EDA) in the realm of cybersecurity. With detailed timestamped logs, the dataset allows for both temporal trend analysis and predictive modeling.

# Dataset Overview

The purpose is to analyze IP traffic, identify potential cybersecurity threats, and assess the effectiveness of protection mechanisms such firewalls and intrusion detection systems (IDS/IPS). It can also be used to identify traffic kinds and assign risk scores, thus improving an organization's capacity to respond proactively to cyber attacks. ## Data Limitations and Improvements
- **Missing Data**: Some fields, such as “IDS/IPS Alerts” and “Proxy Information,” contain missing or incomplete data, which could impact the quality of the analysis.
- **Bias**: The dataset may be biased toward certain types of attacks or networks, which could limit its effectiveness in representing broader real-world cyber threats.
- **Static Data**: While the dataset contains real-time cybersecurity attack data, some parts of it may not reflect the most current trends or attacks in real-time. Regular updates are necessary to maintain its relevance and accuracy.
  
### Suggested Improvements:
- **Data Diversity**: Incorporating more diverse sources, including additional real-world data or threat intelligence feeds, would improve the dataset’s representativeness.
- **Missing Data**: Implementing methods for filling in missing values, such as interpolation or using machine learning models for imputation, could provide a more comprehensive dataset for analysis.
- **Temporal Updates**: Regular updates to reflect ongoing cybersecurity attacks would provide more accurate data for threat detection and model training.

## How the Data Were Created
The dataset was generated by integrating several sources of cybersecurity-related data:
1. **Firewall Logs and IDS/IPS Alerts**: Generated using tools like Snort and Suricata, these logs capture network traffic and alert on suspicious activities.
2. **Packet Sniffers**: Data was captured using tools like Wireshark and tcpdump to analyze the network packets and detect potential malicious activity.
3. **Geo-location Services**: IP addresses are mapped to geographical locations using services like MaxMind to track the origin of network traffic and identify potential sources of cyber attacks.
4. **Real-Time Cybersecurity Attack Data**: Includes data from recent cybersecurity attacks, capturing actual attack incidents, trends, and patterns observed across different network environments.

## Access Rights and Availability
- **Current Access Rights**: There are no immediate privacy or security concerns, but fields like “User Information” or “Geo-location Data” could raise privacy issues if real-world data is integrated or expanded in future iterations.
- **Availability**: The dataset is available for research, development, and educational purposes. Ensuring that sensitive fields like “User Information” are anonymized would allow for wider sharing while maintaining privacy standards.

## 1. Data Acquistion

```python
# Path to your cybersecurity attacks file
pdf_path = "data.pdf"

# Data will be stored in this list
data1 = []

countries = {
    "Russia": ["russia", "russian"],
    "China": ["china", "chinese"],
    "Iran": ["iran", "iranian"],
    "India": ["india", "indian"],
    "Ukraine": ["ukraine", "ukrainian"],
    "United States": ["us", "usa", "united states", "american"],
    "Australia": ["australia", "australian"],
    "North Korea": ["north korea", "north korean"],
    "South Korea": ["south korea", "south korean"],
    "Germany": ["germany", "german"],
    "UK": ["uk", "britain", "british", "united kingdom"],
    "Canada": ["canada", "canadian"],
    "Japan": ["japan", "japanese"],
    "Unknown": [] #can add more countries to this list for future implementation
}

attack_types = {
    "DDoS": ["ddos", "denial of service","massive cyberattack","disturbed","disrupted"],
    "Ransomware": ["ransomware"],
    "Data Leak": ["data leak", "leaked data", "data breach","leak", "stolen", "exposed"],
    "Espionage": ["espionage", "spy", "spying","state-sponsored", "reconnaissance"],
    "Malware": ["malware", "virus", "trojan", "worm"],
    "Phishing": ["phishing", "spear phishing", "social engineering"],
    "Intrusion": ["intrusion", "unauthorized access", "breach","broke into", "hacked"],
    "Backdoor": ["backdoor"],
    "Supply Chain Attack": ["supply chain"],
    "Hacktivism": ["pro-democracy", "hacktivists", "ideological"],
    "Infrastructure Attacks": ["power grid", "utility plant", "infrastructure"],
    "Unknown": [] #Can add more Cybersecurity incidents or attacks to the list for future implementation
}

severity_levels = {
    "High": ["critical", "severe", "major", "significant"],
    "Medium": ["moderate", "notable", "medium"],
    "Low": ["low", "minor", "minimal"] #Add more vocabulary to the list to get exact severity levels from any description
}
with pdfplumber.open(pdf_path) as pdf:
    for page in pdf.pages:
        
        text = page.extract_text()
        lines = text.split("\n")
        
        for line in lines:
            if ":" in line:  
               
                parts = line.split(":", 1)
                timestamp = parts[0].strip()
                details = parts[1].strip().lower()
                country = "Unknown"
                for key, keywords in countries.items():
                    if any(keyword in details for keyword in keywords):
                        country = key
                        break
                attack_type = "Unknown"
                for key, keywords in attack_types.items():
                    if any(keyword in details for keyword in keywords):
                        attack_type = key
                        break
                severity = "Low"  
                for key, keywords in severity_levels.items():
                    if any(keyword in details for keyword in keywords):
                        severity = key
                        break
                data1.append([timestamp, country, details, attack_type, severity])

columns = ["Timestamp", "Geo-location Data", "User Information", "Attack Type", "Severity Level"]
df = pd.DataFrame(data1, columns=columns)
output_path = "Cyberattack_Dataset2.xlsx"
df.to_excel(output_path, index=False)
print(f"Data has been successfully saved to this {output_path} file")
```

## Dataset 2 Cleaning

```python
print(data.columns)
print(df.columns)
```

```python
import pandas as pd

# Load Dataset 1 and Dataset 2
dataset_1_path = "cybersecurity_attacks.csv"  # Path to your Dataset 1 CSV file
dataset_2_path = "Cyberattack_Dataset2.xlsx"  # Path to your Dataset 2 Excel file

# Read the datasets
df1 = pd.read_csv(dataset_1_path)
df2 = pd.read_excel(dataset_2_path)

# Print column names for verification
print("Dataset 1 Columns:", df1.columns)
print("Dataset 2 Columns:", df2.columns)

# Cleaning Dataset 2
# Convert timestamp to a standard datetime format
df2['Timestamp'] = pd.to_datetime(df2['Timestamp'], errors='coerce')

# Fill missing values
df2.fillna({
    "Timestamp": "Unknown",
    "Geo-location": "Unknown",
    "User Info": "Unknown",
    "Attack Type": "Unknown",
    "Severity": "Low",
}, inplace=True)

# Standardize text columns in Dataset 2
text_columns = ['Geo-location Data', 'User Information', 'Attack Type', 'Severity Level']
for col in text_columns:
    df2[col] = df2[col].str.lower().str.strip()

# Drop duplicates in Dataset 2
df2 = df2.drop_duplicates()

# Cleaning Dataset 1
# Convert 'Timestamp' to a standard datetime format
df1['Timestamp'] = pd.to_datetime(df1['Timestamp'], errors='coerce')


# Add missing columns in Dataset 2 to match Dataset 1
for col in df1.columns:
    if col not in df2.columns:
        df2[col] = pd.NA  # Add missing columns as NaN

# Add missing columns in Dataset 1 to match Dataset 2
for col in df2.columns:
    if col not in df1.columns:
        df1[col] = pd.NA  # Add missing columns as NaN

# Ensure all columns are in the same order
df2 = df2[df1.columns]

# Combine the datasets by appending Dataset 2 to Dataset 1
combined_df = pd.concat([df1, df2], ignore_index=True)

# Save the combined dataset
output_path = "Combined_Cybersecurity_Dataset.csv"
combined_df.to_csv(output_path, index=False)
print(f"Combined dataset saved to {output_path}")
```

Cleaning and Preprocessing combined dataset

Now the dataset contains 40,154 rows and 25 columns, with several missing values

```python


# Fill missing numerical values with -1 (or optionally mean/median)
numerical_columns = combined_df.select_dtypes(include=['float64', 'int64']).columns
combined_df[numerical_columns] = combined_df[numerical_columns].fillna(-1)

# 2. Data Type Corrections
# Convert 'Timestamp' to datetime format
combined_df['Timestamp'] = pd.to_datetime(combined_df['Timestamp'], errors='coerce')

combined_df[numerical_columns] = combined_df[numerical_columns].apply(pd.to_numeric, errors='coerce')
data=combined_df
data['year'] = data['Timestamp'].dt.year
data['month'] = data['Timestamp'].dt.month
# Clean 'Severity Level' column
# Standardize to 'low', 'medium', 'high' and replace anything else with 'unknown'
def clean_severity_level(level):
    if pd.isna(level):
        return 'unknown'
    level = str(level).strip().lower()
    if level in ['high', 'medium', 'low']:
        return level
    else:
        return 'unknown'

data['Severity Level'] = data['Severity Level'].apply(clean_severity_level)
data_cleaned=data


```

## 3. Data Analysis
**1. Top 10 Locations with Most Cyberattacks**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Top 10 locations with the highest number of cyberattacks
plt.figure(figsize=(12, 6))
geo_attack_counts = data_cleaned['Geo-location Data'].value_counts().head(10)
sns.barplot(x=geo_attack_counts.values, y=geo_attack_counts.index, palette='viridis')
plt.title('Top 10 Locations with Most Cyberattacks')
plt.xlabel('Number of Attacks')
plt.ylabel('Location')
plt.tight_layout()
plt.show()

```

The bar chart above shows the top 10 locations with the highest number of cyberattacks. Russia, China, and unknown locations are the most targeted, followed by various locations in India. There could be several reasons why there might be an "unknown" location for cyberattacks and the first reason would be due to a sophisticated attacker. Highly skilled hackers often use techniques to mask, making it difficult to trace. Attackers can also use networks of compromised computers (botnets) located in various countries to launch attacks as well as use proxy servers and VPNS. Therefore, the category of "unknown" reflects the limitations in tracking cyberattacks and shows the sophistication of modern cyber threats. 

**2. Cyberattack Trends Over Time**

```python
# Trends over time
plt.figure(figsize=(12, 6))
temporal_trends = data_cleaned.groupby(['year', 'month']).size().reset_index(name='Counts')
temporal_trends['date'] = pd.to_datetime(temporal_trends[['year', 'month']].assign(day=1))
sns.lineplot(x='date', y='Counts', data=temporal_trends, marker='o')
plt.title('Cyberattack Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

The line chart above illustrates the trend of cyberattacks over time. The data shows a significant increase in attacks from 2016 to 2020, with a peak in 2023. However, there is a sharp decline in attacks in 2024. This sharp decline could be attributed to factors such as heightened cybersecurity awareness, improved security technologies, economic shifts, and increased law enforcement efforts. 

**3. Distribution of Severity Levels**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of severity levels
plt.figure(figsize=(10, 6))
severity_counts = data_cleaned['Severity Level'].value_counts()
sns.barplot(x=severity_counts.index, y=severity_counts.values, palette='Set3')
plt.ylim(13300, 13500)
plt.title('Distribution of Severity Levels')
plt.xlabel('Severity Level')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

The bar chart above shows the distribution of severity levels for the cyberattacks. A severity level refers to the seriousness or potential impact of a cyberattack. Factors include data loss, system disruption, financial loss, and reputational damage. A majority of the attacks are classified as "medium" severity, followed by "high" severity. The "low" severity attacks represent the smallest proportion. This bar chart suggests that a significant portion of cyberattacks are moderate in their impact, while a smaller number are considered high-risk.
